# FlyEye Tutorial 1: Analyzing a Single Imaginal Disc

Before working through this tutorial we recommend first familiarizing yourself with the way that FlyEye organizes experimental data into ``Cells``, ``Disc``, and ``Experiment`` objects. These objects and the corresponding data hierarchy are detailed in the getting started section guide on the [FlyEye website](https://www.sbernasek.com/flyeye/start.html#organizing-data), where you'll also find a few high level [usage examples](https://www.sbernasek.com/flyeye/usage.html).


## 1. Creating a Disc from a Silhouette file

We can create a ``Disc`` instance directly from a silhouette file using its ``.from_silhouette(...)`` method. You may use the optional argument ``normalization`` to specify a fluorescence channel used to normalize the intensity of the other fluorescence channels. We typically normalize by the nuclear stain, which in this case is a red His-RFP that resides in the first channel, so ``normalization='ch0'``. 

Loading and processing all of the silhouette data can take some time, so FlyEye automatically caches the result within the corresponding silhouette file so it can be loaded quickly next time. To circumvent this behavior and explicitly update the cache, specify ``recompile=True``.

```python
from flyeye.data import Disc

# specify where the silhouette file lives
SILHOUETTE_FILE_PATH = 'data/disc1.silhouette'

# create a flyeye.Disc instance from a silhouette file
disc = Disc.from_silhouette(
    SILHOUETTE_FILE_PATH, 
    normalization='ch0', 
    recompile=True
)
```

## 2. Exploring the data within a Disc

The data within a ``Disc`` object is stored in a pandas DataFrame, which you can access via the ``.data`` attribute. In this table, each row corresponds to a labeled nuclear contour (measurement), while the columns correspond to various properties of that contour.

These properties include:

 - ``segment_id``: unique ID of the individual contour in Silhouette
 - ``segment``: list of [X,Y] coordinates defining the contour
 - ``centroid_x`` and ``centroid_y``: contour's mean position in the image
 - ``pixel_count``: number of pixels within the contour
 - ``layer``: z-stack layer in which the contour resides
 - ``label``: label assigned in Silhouette
 - ``chX``: mean channel X intensity among the contour's pixels
 - ``chX_std``: std. dev. of channel X intensity among the contour's pixels
 - ``t``: estimated developmental age (calculated during ``Disc`` instantiation)
 
If you specified a normalization channel, you will also see:
 - ``chX_normalized``: normalized mean channel X intensity among the contour's pixels
 - ``chX_normalized_trend``: moving average of normalized channel X intensity 
 - ``chX_normalized_residuals``: absolute deviation about the moving average of normalized channel X intensity 
 - ``chX_normalized_flux``: relative deviation about the moving average of normalized channel X intensity*
 - ``chX_normalized_flux_raw``: relative deviation about the moving average of channel X intensity

```python
disc.data
```

## Filtering by cell type

You can use the ``.select_cell_type(...)`` method to extract one or more specific cell types from the disc. The available cell types are all of those you defined when annotating the disc in ``Silhouette``. For instance, in this example disc we've labeled progenitors as ``pre`` and the R1-8 cell fates as ``r1 ... r8``.

Provide a string for a single cell type, or a list of strings for multiple cell types. 

```python
# selecting progenitors
progenitors = disc.select_cell_type('pre')

# selecting r8s
r8s = disc.select_cell_type('r8')

# selecting all R cells
rcells = disc.select_cell_type(['r1','r2','r3','r4','r5','r6','r7','r8'])
```

These operations all return a ``flyeye.data.Cells`` instance:

```python
print(type(progenitors))
```

A ``Disc`` instance is just a special type of ``Cells`` instance - so most methods and attributes of ``Disc`` objects are also available for ``Cells``, including a ``Cells.data`` table limited to the corresponding records.

```python
progenitors.data
```

## Filtering data by developmental time

Similarly, we can select cells by their developmental age or position within the image using the ``.select_by_position(...)`` method. The arguments of which (tmin, tmax, xmin, xmax, etc.) define the bounds of your selection.

```python
tmin, tmax = 0, 5 # in hours

cells = disc.select_by_position(tmin=tmin, tmax=tmax)
```

Note that applying these methods to ``Cells`` objects just returns another ``Cells`` object, so we can readily dot chain them together.

```python
disc.select_cell_type('r8').select_by_position(tmin=tmin, tmax=tmax).data[['label', 't']]
```

## Plotting expression dynamics

We can plot any column in the ``Cells.data`` table versus developmental time using the ``.plot_dynamics(...)`` method. Use the first argument to specify which channel you'd like to plot. In the example below we've specific ``channel='ch1_normalized'`` to plot DAPI-normalized PntGFP levels.

The ``.plot_dynamics()`` method offers a broad range of customizability as detailed in the [documentation](https://www.sbernasek.com/flyeye/modules/data.html#flyeye.data.cells.Cells.plot_dynamics). A handful of basic examples are shown below.

```python
# plot DAPI-normalized PntGFP dynamics
ax = progenitors.plot_dynamics(
    channel='ch1_normalized', 
    scatter=True,
    average=True,
    line_kw={'color':'black'},
    marker_kw={'color': 'pink', 's': 5}
)
_ = ax.set_ylabel('Pnt (a.u.)')
```

The ``ax`` argument supports passing in an external ``matplotlib.AxesSubplot`` instance. This is a convenient way to both customize the figure as a whole, and to add multiple different elements to the figure.

```python
import matplotlib.pyplot as plt

# create a figure
fig, ax = plt.subplots(figsize=(4, 2.5))
ax.set_xlabel('Time (h)')
ax.set_ylabel('Pnt (a.u.)')

# plot the moving average normalized Pnt-GFP dynamics for progenitors
disc.select_cell_type('pre').plot_dynamics(
    channel='ch1_normalized', 
    scatter=False, 
    average=True, 
    ax=ax, 
    line_kw=dict(label='Progenitors')
)

# overlay the the normalized Pnt-GFP measurements for R8 cells in green
disc.select_cell_type('r8').plot_dynamics(
    channel='ch1_normalized', 
    scatter=True, 
    average=False, 
    ax=ax, 
    marker_kw={'color': 'green', 's': 5, 'label':'R8'}
)

# overlay the the normalized Pnt-GFP measurements for R1 and R6 cells in red
disc.select_cell_type(['r1','r6']).plot_dynamics(
    channel='ch1_normalized', 
    scatter=True, 
    average=False, 
    ax=ax, 
    marker_kw={'color': 'red', 's': 5, 'label':'R1/R6'}
)

ax.legend(loc=0)
```

You may also generate 2D scatterplots using the ``.scatterplot()`` method, which takes the X and Y dimensions as its first two arguments.

```python
# create a figure
fig, ax = plt.subplots(figsize=(3,3))

# plot Pnt-GFP vs Anti-Yan for several cell types
disc.select_cell_type('pre').scatterplot('ch2_normalized','ch1_normalized', ax=ax)
disc.select_cell_type(['r1','r6']).scatterplot('ch2_normalized','ch1_normalized', color='red', s=10, ax=ax)

ax.set_xlabel('Yan (a.u.)')
ax.set_ylabel('Pnt (a.u.)')
```
